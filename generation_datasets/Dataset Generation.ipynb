## Обрабатываем снимки Calc-Test

### Просмотр описания снимков

```python
### Просмотрим данные по снимкам в Calc-Test
data=pd.read_csv('calc_case_description_test_set.csv')

## Отбираем только с видом MLO
data_CC=data.loc[data['image view']=='MLO'].reset_index().drop(columns=['index'])
data_CC.head()
```

### Конвертируем и сохраняем снимки в нужных папках с классами 

```python
import pydicom
import matplotlib.pyplot as plt
import scipy.misc
from skimage import io, transform, feature
from PIL import Image
import os
import re
import numpy as np
```

```python
## Функция для конвертации снимка из DICOM в PNG
def convert_to_png(file, category, number):
    ds = pydicom.dcmread(file)

    shape = ds.pixel_array.shape

    # Convert to float to avoid overflow or underflow losses.
    image_2d = ds.pixel_array.astype(float)

    # Rescaling grey scale between 0-255
    image_2d_scaled = (np.maximum(image_2d,0) / image_2d.max()) * 255.0

    # Convert to uint
    image_2d_scaled = np.uint8(image_2d_scaled)

    # Write the PNG file
    path_file_save='F:\\dataset\\test\\'+str(category)+'\\'+str(number)+str('.png')
    with open(path_file_save, 'wb') as png_file:
        w = png.Writer(shape[1], shape[0], greyscale=True)
        w.write(png_file, image_2d_scaled)
```

```python
for i in range(len(data_CC)):
    path=str('F:\\CBIS-DDSM\\Calc-Test_')+str(data_CC['patient_id'][i])+str('_')+str(data_CC['left or right breast'][i])+str('_')+str(data_CC['image view'][i])
    k=0
    for root, dirs, files in os.walk(path, topdown=True):
        k+=1
        if k==3:
            path_file=str(root)+str('\\')+str(files[0])
            file = path_file.replace('\\', '\\\\')
            category = str(data_CC['assessment'][i])
            number = 'calc_'+str(i)
            convert_to_png(file, category, number)
            k=0
            if i%5==0:
                print('Сохранено '+str(i)+' снимков из '+str(len(data_CC)))
```
